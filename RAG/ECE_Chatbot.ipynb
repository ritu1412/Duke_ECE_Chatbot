```python
from transformers import AutoTokenizer, AutoModelForCausalLM,pipeline
from langchain_community.embeddings import OpenAIEmbeddings
from langchain.schema.runnable import RunnablePassthrough
from langchain_community.vectorstores import Chroma
from langchain.prompts import PromptTemplate
from langchain import HuggingFacePipeline
from langchain.chains import LLMChain
from dotenv import load_dotenv

import gradio as gr
import chromadb
import torch
import os
```

```python

hf_token = os.getenv("HF_READ_TOKEN")
openai_token = os.getenv("OPENAI_TOKEN")
```

```python

device = "cuda" if torch.cuda.is_available() else "cpu"

model_name='mistralai/Mistral-7B-Instruct-v0.2'
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True,token=hf_token)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

model = AutoModelForCausalLM.from_pretrained(
    model_name,token = hf_token,trust_remote_code=True)
model.to(device)

text_generation_pipeline = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    temperature=0.2,
    repetition_penalty=1.1,
    return_full_text=True,
    max_new_tokens=1000,
    pad_token_id=tokenizer.eos_token_id,
    token = hf_token
)

mistral_llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
```

```python
## use a smaller model
device = "cuda" if torch.cuda.is_available() else "cpu"

model_name="microsoft/Phi-3-mini-128k-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True,token=hf_token)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

model = AutoModelForCausalLM.from_pretrained(
    model_name,token = hf_token,trust_remote_code=True)
model.to(device)

text_generation_pipeline = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    temperature=0.2,
    repetition_penalty=1.1,
    return_full_text=True,
    max_new_tokens=1000,
    pad_token_id=tokenizer.eos_token_id,
    token = hf_token
)

phi_llm = HuggingFacePipeline(pipeline=text_generation_pipeline)

```

```python
def get_conversational_chain(langchain_chroma):
    prompt_template = """
    ### [INST] Instruction: Answer the question based on your knowledge. Here is context to help:

    {context}

    ### QUESTION:
    {question} [/INST]
    """

    # Create prompt from prompt template
    prompt = PromptTemplate(
        input_variables=["context", "question"],
        template=prompt_template,
    )

    # Create llm chain
    llm_chain = LLMChain(llm=mistral_llm, prompt=prompt)

    rag_chain = (
    {"context": langchain_chroma.as_retriever(), "question": RunnablePassthrough()}
        | llm_chain
    )
    return rag_chain

# get the vectorstore
client = chromadb.PersistentClient(path="DB/")
embeddings = OpenAIEmbeddings(openai_api_key=openai_token)
langchain_chroma = Chroma(client=client,collection_name='ECE_DATA',
                    embedding_function=embeddings)
conversational_chain = get_conversational_chain(langchain_chroma)


def ECE_chatbot(question):
  bot_message = "Default message"
  bot_message = conversational_chain.invoke(question)['text']
  index = bot_message.find('[/INST]')
  return bot_message[index+9:]

```

```python
def main():
    demo = gr.Blocks()

    chatbot = gr.Interface(
        fn = ECE_chatbot,
        inputs=["textbox"],
        outputs=["textbox"],
        title="ECEBot"
    )

    with demo:
        gr.TabbedInterface(
            [chatbot],
            ["Chatbot"],
        )

    demo.launch(share=True,debug=True)

if __name__ == "__main__":
    main()
```

```python
questions = [
        "Who is the Masters Program Coordinator?",
        "How do you contact the Admissions and Recruiting team",
        "How many credits do you need to graduate with a Masters Degree in ECE?",
        "What are criterias for undergraduates to graduate with distinction?",
        "What are some research areas under quantum computing?",
        "Who is the director of graduate studies?",
        "What are some of the available tracks in ECE?",
        "What are some courses taught by Professor Lawrence Carin?",
        "Name a few key courses in the software engineering track?",
        "How many credits do I have to get just through graduate level ECE courses",
        "Who is the senior program coordinator?",
        "What is Duke university's national ranking?",
        "What is the course code for Introduction to Digital Communication Systems",
        "What is the masters program email id?"
    ]
ground_truth = ["Krista Turner",
        "Send an email to this email id: pratt_masters@duke.edu",
        "Credits required to graduate are 30 credits",
        "Students must maintain a 3.5 cumulative GPA, sucessfully complete a faculty-supervised Independent Study Project in their senior year with significant Electrical and Computer Engineering accomplishment and submit a formal written report and defend it in an oral presentation before a committee of faculty members",
        "Some of the research areas in quantum computing include Quantum error correction,Quantum control, Quantum computer architecture, Ion trapping, Spectroscopy of sympathetically cooled molecular ions, Quantum information, Novel photonic devices, Quantum thermodynamics, Quantum resource theory",
        "Michael E. Gehm",
        "Software Engineering, Computer Hardware Engineering, Machine Learning/ Big data, Quantum Software/Hardware and Semiconductor technology",
        "ECE 899: Special Readings in Electrical Engineering,ECE 891: Internship,ECE 590: Advanced Topics in Electrical and Computer Engineering",
        "ECE 551D: Programming Data Structures and Algorithms in C++, ECE 550D: Fundamentals of Computer Systems and Engineering,ECE 651: Software Engineering,ECE 650: Systems Programming and Engineering",
        "Minimum of 12 units of graduate level ECE courses",
        "Kevyn Light",
        "12 national ranking",
        "ECE 483",
        "ecemastersteam@duke.edu"
    ]
```

```python
contexts = [[langchain_chroma.similarity_search(query=question)[0].page_content] for question in questions]
```

```python
answers = [ECE_chatbot(question) for question in questions]
```

```python
len(answers)
```

```python

from datasets import Dataset
import os
from ragas import evaluate
from ragas.metrics import answer_correctness,context_precision,context_recall,context_relevancy,faithfulness, answer_relevancy

os.environ["OPENAI_API_KEY"] = openai_token

data_samples = {
    'question': questions,
    'answer': answers,
    'contexts': contexts,
    'ground_truth':ground_truth
}

dataset = Dataset.from_dict(data_samples)

score = evaluate(dataset,metrics=[answer_relevancy,answer_correctness,context_precision,context_relevancy,faithfulness])
score.to_pandas()
```

```python
phi_results = score.to_pandas()
```

```python
phi_results.to_csv('phi_eval.csv')
```

```python
print('Average answer relevancy:', phi_results['answer_relevancy'].mean())
print('Average context precision:', phi_results['context_precision'].mean())
print('Average answer faithfulness:', phi_results['faithfulness'].mean())
print('Average answer correctness:',phi_results['answer_correctness'].mean())
```
